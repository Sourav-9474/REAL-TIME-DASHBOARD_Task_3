```python
import requests
import json
from datetime import datetime

url = "https://api.powerbi.com/beta/10bf1100-ea26-4b18-85e6-4ab71af8340d/datasets/44dd5d20-aa79-45d2-858f-d10faa8a236b/rows?experience=power-bi&key=pk0%2BoazE5kxNomDIPAME5Pvv4uSciX7wmZhKyvzrCNASF9wa8z1U6LHrq8p9MY%2Fd3PRJIis6%2FsWSGAiG0lo%2BUQ%3D%3D"

headers = {
    "Content-Type": "application/json"
}

# List of records to send
data_records = [
    {"Temperature": 98.6, "City": "AAAAA555555"},
    {"Temperature": 97.9, "City": "BBBBB222222"},
    {"Temperature": 99.1, "City": "CCCCC333333"},
    {"Temperature": 100.2, "City": "DDDDD444444"}
]

# Attach current datetime to each record and send in one batch
payload = []
for record in data_records:
    payload.append({
        "Temperature": record["Temperature"],
        "City": record["City"],
        "Date Time": datetime.utcnow().isoformat() + "Z"
    })

# Send all records in one POST request
response = requests.post(url, headers=headers, data=json.dumps(payload))

# Output response
print("Status Code:", response.status_code)
print("Response Body:", response.text)
```
